```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 5
b = 4
c = 3
nsamples = 100

x = np.linspace(-4,4,nsamples)

y = a*x**2 + b*x + c + 20*np.random.random(nsamples)
```

```python
plt.rcParams['figure.figsize'] = (10, 6)
plt.plot(x,y,'r*')
plt.xlabel('Samples')
plt.ylabel('Predictions')
plt.show()
```

```python
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X = x.reshape(-1,1)
```

```python
lregr = linear_model.LinearRegression()

lregr.fit(X,y)

yp = lregr.predict(X)

# The mean squared error
print('MSE:',mean_squared_error(y, yp))
print('R2:',r2_score(y, yp))
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly_reg = PolynomialFeatures(degree = 2)
X_poly = poly_reg.fit_transform(X)
lin_reg = LinearRegression()
lin_reg.fit(X_poly, y)
y2 = lin_reg.predict(X_poly)

poly_reg = PolynomialFeatures(degree = 20)
X_poly = poly_reg.fit_transform(X)
lin_reg = LinearRegression()
lin_reg.fit(X_poly, y)
ylin = lin_reg.predict(X_poly)

print('MSE 2:',mean_squared_error(y, y2))
print('R2 2:',r2_score(y, y2))

print('MSE 20:',mean_squared_error(y, ylin))
print('R2 20:',r2_score(y, ylin))
```

```python
plt.plot(X,y,'r*',label='Real')
plt.plot(X,yp,label='Linear')
plt.plot(X,y2,label='x2')
plt.plot(X,ylin,label='x20')
plt.legend(loc=1)

#plt.xlabel('Samples')
#plt.ylabel('Predictions')
plt.show()
```
